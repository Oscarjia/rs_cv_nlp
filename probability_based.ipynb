```
import re
import jieba
#only keep Chinese/English characters and numbers
cop = re.compile("[^\u4e00-\u9fa5a-zA-Z0-9]")
def pure(contents):
  return ''.join(cop.sub('',item) for item in contents)
```

```
pure(contents[:100])
```

```
#tokens 
def tokens(pure):
  return list(jieba.cut(pure))
```

```
len(contents)
```

```
MAX_LENGH=100

Tokens=tokens(pure(contents[:MAX_LENGH]))
```

```
def get_one_gram_count(word):
    if word in words_count:return words_count[word]
    else:
      return words_count.most_common()[-1][-1]

def get_two_gram_count(word):
    if word in two_gram_words_count:return two_gram_words_count[word]
    else:
      return two_gram_words_count.most_common()[-1][-1]

def get_gram_count(word,wc):
    if word in wc:return wc[word]
    else:
      return wc.most_common()[-1][-1]
```

```
def two_gram_model(sentence):
  twotokens=tokens(pure(sentence))
  probability=1
  for i in range(len(twotokens)-1):
    word=twotokens[i]
    nextword=twotokens[i+1]
    prob=get_gram_count(word+nextword,two_gram_words_count)/get_gram_count(nextword,words_count)
    probability*=prob
  return probability
```

```
two_gram_model('小米开发大学')
```
